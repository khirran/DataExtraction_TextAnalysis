```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import os
import codecs
import pyphen
import nltk
nltk.download('cmudict')
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.stem import PorterStemmer
```

```python
dataset = pd.read_csv('/content/Input.xlsx - Sheet1 (1).csv', quoting = 3)
```

```python
files = os.listdir('custom_stopwords')
custom_stopwords = []
for i in files:
  print(i)
```

```python
files = os.listdir('custom_stopwords')
custom_stopwords = []
for i in files:
  f = open('custom_stopwords/'+i)
  try:
    for j in f.readlines():
      custom_stopwords.append(j.strip().lower())
  except:
    print('error',i)
custom_stopwords = set(custom_stopwords)
print(custom_stopwords)
```

```python
sent_count = []
con = []
for i in dataset.iloc[:,-1]:
  r = requests.get(i)
  soup = BeautifulSoup(r.content,"html.parser")
  try:
    title = soup.title.text
    article = soup.article.find_all('p')
    content = title.lower()
    for j in article:
      content+=j.text.lower()
    sent = sent_tokenize(content)
    sent_count.append(len(sent))

    content = word_tokenize(content)
    for k in list(custom_stopwords):
      try:
        content.remove(k)
      except:
        pass
    con.append(content)
  except:
    pass

```

```python
for i in con:
  print(i)
```

```python
ccon = []
word_count = []
punctuators = '''!()-[]{};:'"\,<>./?@#$%^&*_~|'''
for i in con:
  cccon = []
  cnt = 0
  for j in i:
    if j not in punctuators:
      cccon.append(j)
      cnt+=1
  ccon.append(cccon)
  word_count.append(cnt)
con = ccon


```

```python
posfile = open('positive-words.txt','r')
posw = []
for pos in posfile:
  posw.append(pos.strip())
print(posw)
```

```python
#positive score
allpos_scores =[]
for i in con:
  pos_score = 0
  for j in i:
    if j in posw:
      pos_score +=1
  allpos_scores.append(pos_score)
```

```python
negfile = codecs.open('negative-words.txt', 'r', encoding='utf-8',errors='ignore')
negw = []
for neg in negfile.readlines():
  negw.append(neg.strip())
print(negw)
```

```python
#negative scores
allneg_scores =[]
for i in con:
  neg_score = 0
  for j in i:
    if j in negw:
      neg_score +=1
  allneg_scores.append(neg_score)
```

```python
#polarity score
allpol_scores = []
for i in range(len(con)):
  pol_score = (allpos_scores[i] - allneg_scores[i])/((allpos_scores[i] + allneg_scores[i])+0.000001)
  allpol_scores.append(pol_score)
```

```python
#subjectivity score
allsubj_scores = []
for i in range(len(con)):
  subj_score = (allpos_scores[i] + allneg_scores[i])/(word_count[i]+0.000001)
  allsubj_scores.append(subj_score)
```

```python
#Complex Word Count
dic = pyphen.Pyphen(lang='en')
comp_count = []
for i in con:
  cnt = 0
  for j in i:
    cnt+=(len(dic.inserted(j).split('-'))>2)
  comp_count.append(cnt)
print(comp_count)
```

```python
#Analysis of Readability
asl = [word_count[i]/sent_count[i] for i in range(len(con))]
pcw =  [comp_count[i]/word_count[i] for i in range(len(con))]
fog_index = [0.4 * (asl[i] + pcw[i]) for i in range(len(con))]
print(fog_index)
```

```python
#Average Number of Words Per Sentence
aps = [word_count[i]/sent_count[i] for i in range(len(con))]
```
